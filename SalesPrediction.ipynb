load dataset

```python
dataset = pd.read_csv('SalePrediction.csv')
print(dataset)
```

summarised the data

```python
print(dataset.shape) # no. of rows and coloumns
print(dataset.head(5))
print(dataset.tail(5))
```

segragation of data

```python
X = dataset.iloc[:,:-1].values
X
```

```python
Y = dataset.iloc[:,-1].values
Y
```

splitting dataset into train and test

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size= 0.25, random_state=0)
```

Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
```

Training

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
age = int(input("enter new customer's age"))
sal = int(input("enter new customer's salary"))
newCust = [[age, sal]]
result = model.predict(sc.transform(newCust))
print(result)
if result == 1:
  print("Customer Will buy this product")
else:
  print("Customer won't buy")
```

prediction of the product will get buy or not

```python
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

evaluating model confusion matrix

```python
from sklearn.metrics import confusion_matrix , accuracy_score
from sklearn.metrics import accuracy_score
cm = confusion_matrix(y_test , y_pred)

print ("Confusion Matrix:")
print(cm)

print("Accuracy of the model :{0}% ".format(accuracy_score(y_test, y_pred)*100))
```
